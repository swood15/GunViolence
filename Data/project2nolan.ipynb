```python
#dependencies
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import datetime
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot
```

```python
#read csv
data = pd.read_csv("GVA_data (1).csv")
```

```python
data.head()
```

```python
#create new df with needed info
data2 = data[["Year", "Month", "total_casualties", "n_killed", "n_injured"]].copy()
```

```python
#remove all zero casualty totals and make new index
data2 = data2[data2.total_casualties !=0]

data2 = data2.reset_index()

```

```python
#drop old index column
data2 = data2.drop(columns=["index"])
```

```python
data2.head()
```

```python
#group graph by year 
yeargrouped = data2.groupby("Year")
```

```python
yeargrouped.head()
```

```python
#count total number of each column
totals_df = yeargrouped["total_casualties", "n_killed", "n_injured"].sum()
```

```python
totals_df
```

```python
totals_df = totals_df.reset_index()

```

```python
#drop incomplete years 
totals_df = totals_df.drop([0, 5])
```

```python
totals_df
```

```python
#plotly graph
data = [
    go.Bar(
        x=totals_df['Year'], # assign x as the dataframe column 'x'
        y=totals_df['n_killed'],
        name='Number Killed'
    ),
    go.Bar(
        x=totals_df['Year'],
        y=totals_df['n_injured'],
        name='Number Injured'
    )

]
```

```python
layout = go.Layout(
    
    barmode='stack',
    title='Casualty Comparison of Number Killed vs Number Injured',
    xaxis=dict(
        title='Year',
    ),
    yaxis=dict(
        title='Number of Casualties',
    )    

)

fig = go.Figure(data=data, layout=layout)
```
